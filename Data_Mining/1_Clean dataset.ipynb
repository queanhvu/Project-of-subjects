```python
#@title xoa hang trung lap
cat_df = cat_df.drop_duplicates()
cat_df.info()
```

## Preprocessing

```python
import sklearn
```

```python
from sklearn.preprocessing import OneHotEncoder as OE
```

```python
def onehot(x, codedict):
  if x is None or str(x) == "Unknown" or str(x).strip() == "" or str(x) == "nan":
    return None
  i = codedict.index(x)
  if i < 0 : return None
  else : return i
```

```python
#@title onehot name
cat_df["name"] = cat_df["name"].apply(lambda x: 1 if x is None else 0)
```

```python
#@title onehot sex_upon_outcome
Suo = list(cat_df["sex_upon_outcome"].unique())
cat_df["sex_upon_outcome"] = cat_df['sex_upon_outcome'].apply(lambda x: onehot(x, Suo))
Suo 
```

```python
#@title onehot cat/kitten
name = "Cat/Kitten (outcome)"
CoKn = list(cat_df[name].unique())
cat_df[name] = cat_df[name].apply(lambda x: onehot(x, CoKn))
CoKn
```

```python
#@title age group
name = "age_group"
AgG = list(cat_df[name].unique())
cat_df[name] = cat_df[name].apply(lambda x: onehot(x, AgG))
AgG
```

```python
#@title weekday
name = "outcome_weekday"
weekday = list(cat_df[name].unique())
cat_df[name] = cat_df[name].apply(lambda x: onehot(x, weekday))
weekday
```

```python
#@title breed1
name = "breed1"
breed1 = list(cat_df[name].unique())
cat_df[name] = cat_df[name].apply(lambda x: onehot(x, breed1))
breed1
```

```python
#@title breed2
name = "breed2"
breed2 = list(cat_df[name].unique())
cat_df[name] = cat_df[name].apply(lambda x: onehot(x, breed2))
breed2
```

```python
#@title coat_pattern
name = "coat_pattern"
coatpt = list(cat_df[name].unique())
cat_df[name] = cat_df[name].apply(lambda x: onehot(x, coatpt))
coatpt
```

```python
#@title coat
name = "coat"
coat = list(cat_df[name].unique())
cat_df[name] = cat_df[name].apply(lambda x: onehot(x, coat))
coat
```

```python
#@title color1 + 2
name = "color1"
color1 = list(cat_df[name].unique())
cat_df[name] = cat_df[name].apply(lambda x: onehot(x, color1))


name = "color2"
color2 = list(cat_df[name].unique())
cat_df[name] = cat_df[name].apply(lambda x: onehot(x, color2))
color1, color2
```
